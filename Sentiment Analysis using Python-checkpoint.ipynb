```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame already loaded
data=pd.read_csv("\\Users\\HP\\Downloads\\archive (5).zip")

# Step 1: Count the occurrences of each unique category in 'Category' column
category_counts = data['Category'].value_counts()

# Step 2: Create an explode list with the same length as category_counts
# Set the first element to 0.1 (or another value to highlight) and others to 0
explode = [0.1] + [0] * (len(category_counts) - 1)

# Step 3: Plot the pie chart
fig = plt.figure(figsize=(5, 5))

# Ensure you have enough colors for all categories (add more if needed)
colors = ["skyblue", "pink", "lightgreen", "yellow", "orange", "purple", "brown", "grey"]

# Create the pie chart
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%',
        shadow=True, colors=colors[:len(category_counts)], startangle=45, explode=explode)
plt.title("Category Distribution")
plt.show()

# Optional: Print the category counts to see the breakdown
print(category_counts)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from textblob import TextBlob
data=pd.read_csv("\\Users\\HP\\Downloads\\archive (5).zip")

# Step 1: Sentiment Analysis using TextBlob
def get_sentiment(text):
    analysis = TextBlob(text)
    # Classify as positive (1) if polarity > 0, negative (0) otherwise
    return 1 if analysis.sentiment.polarity > 0 else 0

# Apply the sentiment analysis function to the 'Resume' column
data['label'] = data['Resume'].apply(get_sentiment)

# Step 2: Count the occurrences of Positive (1) and Negative (0)
pos = data[data['label'] == 1]
neg = data[data['label'] == 0]

# Step 3: Prepare data for the pie chart
ck = [pos['label'].count(), neg['label'].count()]

# Step 4: Plot the pie chart
fig = plt.figure(figsize=(5, 5))
colors = ["skyblue", "pink"]  # Colors for Positive and Negative labels

plt.pie(ck, labels=["Positive", "Negative"],
        autopct='%1.1f%%', shadow=True, colors=colors,
        startangle=45, explode=(0, 0.1))  # Exploding the Negative slice
plt.title("Positive vs Negative Distribution")
plt.show()

# Optional: Print the counts
print(f"Positive count: {pos['label'].count()}")
print(f"Negative count: {neg['label'].count()}")
```

```python
import pandas as pd
import re



# Preprocessing function to clean text and extract emojis
def preprocessor(text):
    text = re.sub('<[^>]*>', '', text)  # Remove HTML tags
    emojis = re.findall('(?::|;|=)(?:-)?(?:\)|\(|D|P)', text)  # Find emojis
    text = re.sub('[\W]+', ' ', text.lower()) + ' '.join(emojis).replace('-', '')  # Clean and append emojis
    return text

# Apply the preprocessor to the 'Resume' column
data['processed_text'] = data['Resume'].apply(preprocessor)


print(data[['Category', 'Resume', 'processed_text']])
```

```python

from nltk.stem.porter import PorterStemmer
porter=PorterStemmer()
def tokenizer(text):
        return text.split()
def tokenizer_porter(text):
    return [porter.stem(word) for word in text.split()]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Assuming tokenizer_porter is defined
tfidf = TfidfVectorizer(strip_accents=None, lowercase=False, preprocessor=None, 
                        tokenizer=tokenizer_porter, use_idf=True, norm='l2', smooth_idf=True)

# Ensure the 'label' column exists
y = data['label'].values  # Assuming 'label' column is present in your dataset

# Transform the 'processed_text' column, not 'text'
x = tfidf.fit_transform(data['processed_text'])

# Optionally suppress warnings if desired
import warnings
warnings.filterwarnings("ignore", category=UserWarning, module='sklearn')

# Proceed with further model training or evaluation
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,random_state=1,test_size=0.5,shuffle=False)
```

```python
from sklearn.linear_model import LogisticRegressionCV
clf=LogisticRegressionCV(cv=6,scoring='accuracy',random_state=0,n_jobs=-1,verbose=3,max_iter=500).fit(X_train,y_train)
y_pred = clf.predict(X_test)
from sklearn import metrics

# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
